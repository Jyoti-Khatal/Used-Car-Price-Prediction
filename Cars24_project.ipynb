Univariate Linear Regression
As we can see from heat map that model and selling price is more correlated we'll consider x= model and y as we selling price.

Here we are building Linear Regression model from scratch.

we know the linear regression equation as  y_hat  = w0 + w1x (single feature)



```python
x = df['model'].values
y = df['selling_price'].values
```

```python
u = x.mean()
std = x.std()
x = (x-u)/std #standardization
```

```python
def predict(x,weights):
  y_hat = weights[0] + weights[1]*x
  return y_hat
```

Calculating error Metric MSE

As we know MSE = 1/m summation from 1 to m(Y_pred - yi)**2

```python
def error(x,y,weights):
  m = x.shape[0]
  total_error = 0.0
  for i in range(m):
    y_hat = predict(x[i],weights)
    total_error =+ (y_hat-y[i])**2
  return(total_error/m)
```

Now we'll calculate gradient descent for single feature "model"

```python
def gradient(x,y,weights):
  m = x.shape[0]
  grad = np.zeros((2,))
  for i in range(m):
    xi=x[i]
    y_hat = predict(xi,weights)
    yi= y[i]

    grad[0] += -2*(yi-y_hat)
    grad[1] += -2*(yi-y_hat)*xi
  return grad/m
```

```python
def gradient_descent(x,y,epochs=800, learning_rate=0.1):
   weight = np.zeros((2,))
   error_list = []
   weight_list = []

   for i in range(epochs):
        # Compute grad
        grad = gradient(x,y,weight)
        # print(grad)
        e = error(x,y,weight)

        #Update weights
        weight[0] = weight[0] - learning_rate*grad[0]
        weight[1] = weight[1] - learning_rate*grad[1]
        # Storing the weight values during updates
        weight_list.append((weight[0],weight[1]))
        error_list.append(e)

   return weight,error_list,weight_list
```

```python
weight, error_list, weight_list = gradient_descent(x,y,epochs=800)

# Weights after 800 epochs
print(weight)
```

Multivariant Linear Regression

```python
X = df[df.columns.drop('selling_price')]
Y = df["selling_price"]
```

```python
X = X.to_numpy()
Y = Y.to_numpy()
```

```python
u = np.mean(X,axis=0)
std = np.std(X,axis=0)

X = (X-u)/std
```

```python
ones = np.ones((X.shape[0],1))
X = np.hstack((ones,X))
```

```python
def hypothesis(X,weight):
    return np.dot(X,weight)
```

```python
def error(X,y,weight):
    e = 0.0
    m = X.shape[0]
    y_hat = hypothesis(X,weight)
    e = np.sum((y-y_hat)**2)
    return e/m
```

```python
def gradient(X,y,weight):

    y_hat = hypothesis(X,weight)
    grad = np.dot(X.T,(y_hat - y))
    m = X.shape[0]
    return grad/m

def gradient_descent(X,y,learning_rate = 0.1,epochs=500):

    n = X.shape[1]
    weight = np.zeros((n,))
    error_list = []

    for i in range(epochs):
        e = error(X,y,weight)
        error_list.append(e)

        #Gradient descent
        grad = gradient(X,y,weight)
        weight = weight - learning_rate*grad

    return weight, error_list
```

```python
import time
start = time.time()
weight, error_list = gradient_descent(X,Y)
end = time.time()
print("Time taken is ", end-start)
```

```python
def r2_score(y,y_):
    num = np.sum((y-y_)**2)
    denom = np.sum((y- y.mean())**2)
    score = (1- num/denom)
    return score
```
